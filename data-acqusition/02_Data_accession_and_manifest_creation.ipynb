```python
import numpy as np
import pandas as pd
import glob
```

Primero cargamos la encuesta del American Gut Project extraida de Qiita (study=10317) el 11 de abril de 2022.

```python
survey = pd.read_table('metadata_survey.tsv', low_memory=False)
survey
```

Ahora cargamos la tabla con la información de acceso de todas las lecturas disponibles en el ENA (esta tabla se extrajo también del ENA).

```python
reference = pd.read_table('filereport_read_run_PRJEB11419.tsv')
reference 
```

Ahora, cargamos las tablas que contienen la información del numbre de la muestra asignado en el AGT para cada muestra con su respectiva información de acceso en el ENA, y las vamos a concatenar todas en una sola tabla para facilitar su manipulación. Todas estas tablas fueron obtneidas de Qiita.

```python
accession_files = glob.glob('ebi-experiment-accessions/ebi_experiment_accessions_study_10317_prep_*.tsv')
accessions = [pd.read_table(i, dtype=str) for i in accession_files]
accessions
accession = pd.concat(accessions, ignore_index=True)
accession = accession[accession['experiment_accession'] != 'None']
accession
```

En este paso, unimos todas las tablas mapenado la información relevante de cada tabla y agregamos las columnas necesarias para crear el "manifest file" para poder importar los datos a Qiime2.

```python
metadata = survey.merge(accession, on = 'sample_name')
metadata = metadata.merge(reference, on = 'experiment_accession')
absolute_path = '/hpcfs/home/cursos/biocom4102/Grupo_5/Proyecto/data-acquisition/raw-data'
metadata['sample-id'] = metadata['fastq_ftp'] \
                        .str.replace('ftp.sra.ebi.ac.uk/vol1/fastq/ERR\d\d\d/\d\d\d/ERR\d\d\d\d\d\d\d/', '', regex=True) \
                        .str.replace('.fastq.gz', '', regex=False)
absolute_filepath = ['{0}/{1}.fastq.gz'.format(absolute_path, i) for i in metadata['sample-id']]
metadata['absolute-filepath'] = absolute_filepath
metadata.set_index('sample-id', inplace = True)
metadata
```

Ahora filtramos la tabla para obtener únicamente las observaciones que son de nuestro interés. En este caso, nos importa que de las muestras se haya hecho efectivamente la extracción y secuenciación del ADN, que el individuo sea sano (definido como todo individuo entre el rango de edad de los 18 y los 69 años, con un indice de masa corporal entre 18.5 y 30, que no haya reportado que consumió antibióticos en el el transcurso de un año antes de haberse tomado la muestra, y que no haya reportado tener diabetes ni síndrome del colon irritable), que la muestra sea del intestino humano y que el individuo sea residente de los Estados Unidos.

```python
metadata_filtered = metadata[(metadata['dna_extracted'] == True) & \
                             (metadata['subset_healthy'] == 'true') & \
                             (metadata['bmi_cat'] == 'Normal') & \
                             (metadata['env_package'] == 'human-gut') & \
                             (metadata['country_residence'] == 'United States')].copy()
metadata_filtered
```

Acá definimos todas las columnas que nos interesan y el orden en que queremos que se dispongan.

```python
columns_interest = ['absolute-filepath', 
                    'artificial_sweeteners',
                    'diet_type', 
                    'drinking_water_source', 
                    'env_package', 
                    'exercise_frequency', 
                    'fermented_plant_frequency', 
                    'frozen_dessert_frequency', 
                    'fruit_frequency', 
                    'gluten', 
                    'high_fat_red_meat_frequency',
                    'homecooked_meals_frequency',
                    'meat_eggs_frequency',
                    'milk_cheese_frequency',
                    'milk_substitute_frequency',
                    'multivitamin', 
                    'olive_oil',
                    'one_liter_of_water_a_day_frequency',
                    'other_supplement_frequency',
                    'prepared_meals_frequency', 
                    'probiotic_frequency',
                    'ready_to_eat_meals_frequency', 
                    'red_meat_frequency',  
                    'seafood_frequency',  
                    'sugar_sweetened_drink_frequency', 
                    'sugary_sweets_frequency', 
                    'vegetable_frequency', 
                    'vitamin_b_supplement_frequency', 
                    'vitamin_d_supplement_frequency', 
                    'whole_eggs', 
                    'whole_grain_frequency', 
                    'fastq_ftp']
```

Finalmente, creamos el manifest file y lo guardamos en formato tsv.

```python
metadata_interest_filtered = metadata_filtered.loc[:, columns_interest]
metadata_interest_filtered.to_csv('manifest_file.tsv', sep = '\t')
metadata_interest_filtered
```
